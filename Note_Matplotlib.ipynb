```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Define a function to plot
def plot_function(x):
    y = np.square(x)
    return y

# Define the x-values
x = np.linspace(-5, 5, 100)

# Create the figure and axis
fig, ax = plt.subplots()

# Create the line plot
line, = ax.plot(x, plot_function(x))

# Define the animation function
def animate(frame):
    
    # Update the x and y data of the line
    line.set_data(x[:frame], plot_function(x[:frame]))
    
    # Set the x and y limits of the plot
    ax.set_xlim([-5, 5])
    ax.set_ylim([0, 25])
    
    # Set the plot title
    ax.set_title('Squared Curve')
    
    # Force the plot to update
    plt.draw()
    plt.pause(0.001)
    
    return line,


# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=len(x), interval=50)

# Save the animation as an MP4 file
ani.save('square_curve_animation.mp4', writer='ffmpeg')
```
